    - Priority 2: The "Silent Risk" Group (False Negatives, ~200 users)Profile: Look safe (long tenure) but are leaving anyway.Why: These are dangerous "silent killers" leaving due to service failures or unseen competitor offers.
    - Priority 3: The "Loyal but Risky" Group (False Positives, ~260 users)Profile: "Sticky" customers who look like churners but stay.Why: They are candidates for upselling, not saving.
